# Define RLE Decoder

```python
def rle_decode(mask_rle, shape=(768, 768)):
    #     '''
    #     mask_rle: run-length as string formated (start length)
    #     shape: (height,width) of array to return 
    #     Returns numpy array, 1 - mask, 0 - background
    #     '''
        s = mask_rle.split()
        starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
        starts -= 1
        ends = starts + lengths
        img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
        for lo, hi in zip(starts, ends):
            img[lo:hi] = 1
        return img.reshape(shape).T  # Needed to align to RLE direction
```

# Define ShipDataset inherit torch Dataset

```python
class ShipDataset(Dataset):
    """
    Ship Dataset 
    - notice this Dataset only read images with ships
    """

    def __init__(self, csv_filepath, root_dir,train='train', transform=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        # read csv file to shipInfo
        masks = pd.read_csv(csv_filepath)
        # randomly drop no ship
        if train=="train":
            masks = masks.drop(masks[masks.EncodedPixels.isnull()].sample(70000,random_state=42).index)
        self.shipInfo = masks
        self.train=train
        # read the unique images with ship 
        # Note: images without ship has been filtered
        if self.train=='train':
            self.shipImgNames = self.shipInfo.loc[(pd.isna(self.shipInfo['EncodedPixels'])!=True),'ImageId'].unique()
#             self.shipImgNames = self.shipInfo['ImageId'].unique()
        else:
            self.shipImgNames = self.shipInfo['ImageId'].unique()
        self.root_dir = root_dir
        self.transform = transform
        
        self.img_transform = Compose([
                                    ToTensor(),
#                                     Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
                                    ])
        
    def __len__(self):
        # removed duliplated image names
        return len(self.shipImgNames)

    def __getitem__(self, idx):
        
        # get image
        img_name=self.shipImgNames[idx]
        
        # get image path
        img_path = os.path.join(self.root_dir,img_name)
        
        # import image
        img = io.imread(img_path)
        
        # process image in np.array form
        if self.transform:
            img = self.transform(img)
        
        # process the image data to torch.FloatTensor type
#         img = np.atleast_3d(img).transpose(2, 0, 1).astype(np.float32)
#         img = (img - img.min()) / (img.max() - img.min())
#         img = torch.from_numpy(img).float()
        img = self.img_transform(img)
        
        # masks procesing
        # Take the individual ship masks and create a single mask array for all ships
        # masks = self.shipInfo.loc[self.shipInfo['ImageId'] == img_name, 'EncodedPixels'].tolist()
        
        if self.train=='train':
            masks = self.shipInfo.loc[(self.shipInfo['ImageId'] == img_name), 'EncodedPixels'].tolist()
            img_mask = np.zeros((768, 768))

            # ship number counter && ship mask merge into oen
            ship_num = 0 
            if not pd.isnull(masks[0]):
                ship_num = len(masks) 
                for mask in masks:
                    img_mask += rle_decode(mask)

            img_mask = torch.from_numpy(img_mask).float()
        
            # return ship_num, ship number could be returned but decided not to return
            return img, img_mask
        else:
            return img, img_name
```

# Define Unet downblock & upperblock

```python
class UNET_Down(nn.Module):
    def __init__(self,in_channel,out_channel,kernel_size=3,stride=1,padding=1):
        super().__init__()
        self.max_pool=nn.MaxPool2d(2,stride=2)
        self.conv1=nn.Conv2d(in_channel,out_channel,kernel_size=kernel_size,stride=stride,padding=1)
        self.bn1=nn.BatchNorm2d(out_channel)
        self.conv2=nn.Conv2d(out_channel,out_channel,kernel_size=kernel_size,stride=stride,padding=1)
        self.bn2=nn.BatchNorm2d(out_channel)
    def forward(self,x):
        x=self.max_pool(x)
        x=F.relu(self.conv1(x))
        x=self.bn1(x)
        x=F.relu(self.conv2(x))
        x=self.bn2(x)
        return x

class UNET_Up(nn.Module):
    def __init__(self,in_channel,prev_channel,out_channel,kernel_size=3,stride=1,padding=1):
        super().__init__()
        self.up_sample=nn.UpsamplingBilinear2d(scale_factor=2)
        self.conv1=nn.Conv2d(in_channel+prev_channel,out_channel,kernel_size=kernel_size,stride=stride,padding=padding)
        self.bn1=nn.BatchNorm2d(out_channel)
        self.conv2=nn.Conv2d(out_channel,out_channel,kernel_size=kernel_size,stride=stride,padding=padding)
        self.bn2=nn.BatchNorm2d(out_channel)
    def forward(self,x,prev_layer):
        x=self.up_sample(x)       
        x=torch.cat((x, prev_layer), dim=1)
        x=F.relu(self.conv1(x))
        x=self.bn1(x)
        x=F.relu(self.conv2(x))
        x=self.bn2(x)
        return x  
```

```python
class UNET(nn.Module):
    def __init__(self, in_channel=3):
        super().__init__()
        self.conv1=nn.Sequential(nn.Conv2d(in_channel,32,kernel_size=3,stride=1, padding=1),
                                 nn.ReLU(inplace=True),
                                 nn.BatchNorm2d(32),
                                 nn.Conv2d(32,32,kernel_size=3,stride=1, padding=1),
                                 nn.ReLU(inplace=True),
                                 nn.BatchNorm2d(32)
                                )
        self.down1=UNET_Down(32,64)
        self.down2=UNET_Down(64,128)
        self.down3=UNET_Down(128,256)
        self.down4=UNET_Down(256,512)
        self.up4=UNET_Up(512,256,256)
        self.up3=UNET_Up(256,128,128)
        self.up2=UNET_Up(128,64,64)
        self.up1=UNET_Up(64,32,32)
        self.out=nn.Conv2d(32,1,kernel_size=1,stride=1, padding=0)
                                 
    def forward(self, x):
        out1=self.conv1(x)  # 768 32
        out2=self.down1(out1) # 384 64
        out3=self.down2(out2) # 192 128
        out4=self.down3(out3) # 96 256
        ret=self.down4(out4) # 48 512
        ret=self.up4(ret,out4) #
        ret=self.up3(ret,out3)
        ret=self.up2(ret,out2)
        ret=self.up1(ret,out1)
        ret=self.out(ret)
        ret=F.sigmoid(ret)
        return ret
```

# Pre-Training Definition

```python

USE_GPU = True

dtype = torch.float32 # we will be using float throughout this tutorial

if USE_GPU and torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')

# Constant to control how frequently we print train loss
print_every = 100
print('using device:', device)

```

```python
from tqdm import tqdm
def check_accuracy(model, loader):
    
    model.eval()
    with torch.no_grad():
        loss_avg=0.0
        cnt=0
        for x, y in loader:
            x = x.to(device=device, dtype=dtype)  # move to device, e.g. GPU
            y = y.to(device=device, dtype=dtype)
            scores=model(x)
            loss=F.binary_cross_entropy(scores.squeeze(),y)
            loss_avg+=loss.item()
            cnt+=1
            
    return loss/cnt

def train(model, optimizer, load_train, load_val, epochs=1, lr_start=1e-3,lr_end=1e-4, print_every=100):
    loss_history=[]
    cnt=0
    for m in range(epochs):
        for i,(x, y) in enumerate(tqdm(load_train)):

            model = model.to(device=device)
            model.train()
            x = x.to(device=device, dtype=dtype)  # move to device, e.g. GPU
            y = y.to(device=device, dtype=dtype)
            scores=model(x)
            loss=F.binary_cross_entropy(scores.squeeze(),y)

            optimizer.zero_grad()
            loss.backward()

            optimizer.step()

            if i % print_every == 0:
                loss_history.append(loss.item())
                print('Iteration %d, loss = %.4f' % (i, loss.item()))
                check_accuracy(model,load_val)
    return loss_history
```

```python
from skimage.morphology import binary_opening, disk
from skimage.morphology import label

def rle_encode(img):
    '''
    img: numpy array, 1 - mask, 0 - background
    Returns run length as string formated
    '''
    pixels = img.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return ' '.join(str(x) for x in runs)

def multi_rle_encode(img):
    labels = label(img)
    return [rle_encode(labels==k) for k in np.unique(labels[labels>0])]


def predict(model, load_test, print_every=100):
    out_pred_rows=[]
    for i,(x,name) in enumerate(tqdm(load_test)):
        model = model.to(device=device)
        model.train()
        x = x.to(device=device, dtype=dtype)  # move to device, e.g. GPU
        scores=model(x)
        mask=scores.squeeze().detach().cpu().numpy()
        cur_seg = binary_opening(mask>0.7)
        cur_rles = multi_rle_encode(cur_seg)
        if len(cur_rles)>0:
            for c_rle in cur_rles:
                out_pred_rows += [{'ImageId': name[0], 'EncodedPixels': c_rle}]
        else:
            out_pred_rows += [{'ImageId': name[0], 'EncodedPixels': None}]
            
    return out_pred_rows
```
